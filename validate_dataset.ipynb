- a subdirectory /samples/ containing samples in mp3 format.

```python
from datasets import ClassLabel
import random
import numpy as np
import pandas as pd
from IPython.display import display, HTML, Audio
from datasets import load_dataset
import os
```

```python
# This must be path to a directory containing the data.csv dataset file and a subdir /samples/ 
# containing the samples in data.csv
path_to_dataset = '/content/drive/MyDrive/ASR/custom_test_set/'
csv_path = os.path.join(path_to_dataset, 'data.csv')
samples_dir_path = os.path.join(path_to_dataset, 'samples/')
dataset = load_dataset('csv', data_files=csv_path)
dataset = dataset['train']
```

```python
# Loop over examples to validate them
valid = []
num_samples = 5 #len(dataset)
for idx in range(num_samples):
  df = pd.DataFrame(dataset[[idx]])
  track = Audio(samples_dir_path+dataset[idx]["filename"], autoplay=True)
  display(HTML(df.to_html()))
  display(track)
  accept = input('accept? [y/n] ')
  if accept.lower() == 'y':
    valid.append(idx)
```

```python
# Load initial dataset and filter out unwanted samples.
validated_csv = os.path.join(path_to_dataset,'validated_data.csv')
dataframe = pd.read_csv(csv_path)
dataframe = dataframe.iloc[valid]
dataframe.to_csv(validated_csv, index=False)
```
